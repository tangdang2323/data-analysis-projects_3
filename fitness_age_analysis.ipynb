Поработаем со списком возрастов посетителей фитнес-клубов age_lst

**Задача 1.** Рассчитаем средний возраст по всему списку age_lst

**Задача 2.** Рассчитаем средний возраст для посетителей, которые моложе 50 лет

**Задача 3.** Обернем предыдущий код в функцию, которая на вход принимает список и пороговое значение

**Задача 4.** Рассчитаем среднее значение для всех элементов списка, за исключением тех, которые лежат вне интервала [Q1-1,5*IQR; Q3+1,5*IQR]

**Задача 5.** Обернем предыдущий код в функцию, которая принимает на вход список и рассчитывает средний возраст, исключая из расчётов выбросы

```python
age_lst = [23, 45, 18, 80, 22, 25, 34, 37, 41, 88, 19, 20, 18, 45, 41, 39, 31, 32, 22, 28, 37]
```

```python
#Задача1
avg_age = sum(age_lst)/len(age_lst)
print(round(avg_age,2))
```

```python
#Задача2
summ = 0
cnt = 0
for i in age_lst:
    if i < 50:
        summ += i
        cnt += 1
avg_m_age = summ/cnt
print(round(avg_m_age,2))
```

```python
#Задача3
def avg_calc(lst,thresh):
    summ = 0
    cnt = 0
    for i in lst:
        if i < thresh:
            summ += i
            cnt += 1
    avg_m_age = summ/cnt
    return round(avg_m_age,2)
```

```python
avg_calc(age_lst,50)
```

```python
#Задача4
import numpy as np

q1 = np.percentile(age_lst,25)
q3 = np.percentile(age_lst,75)
iqr = q3 - q1
border_lower = q1 - 1.5 * iqr
border_upper = q3 + 1.5*iqr
summ = 0
cnt = 0 
for i in age_lst:
    if i>=border_lower and i<= border_upper:
        summ += i
        cnt += 1
avg = summ/cnt
print(round(avg,2))
```
